## Question 5 — Loops with multiple lists (nested for loops)

A *nested* loop is simply a `for` loop inside another `for` loop. The inner loop runs from start to end every time the outer loop advances by one step. Here is a tiny worked example:

```python
outer_list = ["a", "b", "c"]
inner_list = [1, 2]

for letter in outer_list:        # outer loop
    for number in inner_list:    # inner loop (runs fully for each letter)
        print(letter, number)
```

This prints `a 1`, `a 2`, `b 1`, `b 2`, `c 1`, `c 2` — six lines in total (3 × 2). You will use the same pattern below. If you would like an additional refresher, see <https://www.w3schools.com/python/gloss_python_for_nested.asp>.
