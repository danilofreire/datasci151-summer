# Flow control: conditional statements 🐍

## Flow control: conditional statements

### If, elif, else

- **Conditional statements** are used to make decisions in programming
- The **`if`** statement runs a block of code if a condition is `True`
- The **`elif`** statement runs a block of code if the first condition is `False` and the second condition is `True`
- The **`else`** statement runs a block of code if all other conditions are `False`
- We use this a lot in data cleaning or data analysis to filter out data!

[![](figures/if.png){width="70%"}](#){data-modal-type="image" data-modal-url="figures/if.png"}

## If statements

- Two parts of an `if`:
  - The **test expression**: `if`, a condition, and `:`
  - The **body**: the code to run, [indented with four spaces](https://www.python.org/dev/peps/pep-0008/#indentation)
- Tip: press `Tab` to indent in your editor
- If the condition is `True`, the body runs
- If the condition is `False`, the body is skipped

- The `if` statement syntax is as follows:

```python
if condition:
    Body
```

- Another example:

```python
# We start by defining a string
any_questions = "yes"

if any_questions == "yes":
    print("More explanation needed")
```

## If/else statements

### When you have two possible outcomes

- The `else` statement is used to execute a block of code if the condition is `False`
- The `else` statement is always used in conjunction with the `if` statement
- The `else` statement syntax is as follows:

```python
if condition:
    Body
else:
    Body
```

- For example, let's say you want to know whether someone can vote in the US
  - First, the person needs to be a US citizen
  - Second, the person needs to be at least 18 years old

```python
age = 22
is_citizen = True   # Try setting this to False to see the `else` branch

if is_citizen and (age >= 18):
    print("This person can vote in the US")
else:
    print("This person cannot vote in the US")
```

## If/elif/else statements

- The `elif` statement **checks the next condition** only when the previous one was `False`
- Useful when you have **more than two outcomes**
- Syntax:

```python
if condition:
    Body
elif condition:
    Body
else:
    Body
```

[![](figures/elif.png){width="70%"}](#){data-modal-type="image" data-modal-url="figures/elif.png"}

## Example

```python
years_in_program = 2

if years_in_program == 1:
    print("This student is a freshman")
elif years_in_program == 2:
    print("This student is a sophomore")
elif years_in_program == 3:
    print("This student is a junior")
else:
    print("This student is a senior")
```

# Combining conditions and logical operators 🤓

# Nested if statements 🪆

## Nested `if` statements

### When one condition depends on another

- An `if` statement can be placed inside another `if` (or `elif`/`else`) statement. This is called **nesting**!
- Used for more granular decision-making where one check depends on the outcome of a previous one
- Caution: **nesting** too deeply (more than 2-3 levels) makes code hard to read and debug

- Syntax example:

```python
if condition1:
    # Outer if body
    print("Condition 1 is True")
    if condition2:
        # Inner if body 
        # Executes if condition1 AND condition2 are True
        print("Condition 2 is also True")
    else:
        # Inner else body
        # Executes if condition1 is True BUT condition2 is False
        print("Condition 2 is False")
else:
    # Outer else body
    print("Condition 1 is False")
```

The path is like a tree: 1. Check `condition1` (outer `if`). 2. If `True`, check `condition2` (inner `if`). 3. If `False`, go to the inner `else`. 4. If `condition1` is `False`, go to the outer `else`

## Nested `if` example: ticket pricing

Scenario: ticket price depends on age group, and then on the day of the week for adults

```python
age = 30
day = "Saturday" # Try "Monday"
price = 0

if age < 18: # Child
    price = 10
    category = "Child"
elif 18 <= age < 65: # Adult (Python lets you chain comparisons)
    category = "Adult"
    if day == "Saturday" or day == "Sunday":
        price = 20 # Weekend adult price
    else:
        price = 15 # Weekday adult price
else: # Senior
    price = 12
    category = "Senior"

print(f"Category: {category}, Day: {day}, Price: ${price}")
```

# Conditional expressions (ternary operator) ✨

## Conditional expressions (ternary operator)

### A concise way to assign based on condition

- For simple `if/else` that just **picks a value**, Python has a **one-liner**
- Also called a **ternary operator** (three parts joined together)
- Syntax: `value_if_true if condition else value_if_false`
  - If `condition` is `True`, the whole expression evaluates to `value_if_true`
  - If `condition` is `False`, the whole expression evaluates to `value_if_false`

## Conditional expression examples

Standard `if/else` for assignment:

```python
age = 20
status = ""
if age >= 18:
    status = "Adult"
else:
    status = "Minor"
print(f"Status (standard if/else): {status}")
```

Using a conditional expression:

```python
age = 20 # Same age
status_ternary = "Adult" if age >= 18 else "Minor"
print(f"Status (ternary operator): {status_ternary}")
```

Another example, setting a discount:

```python
is_member = True
# is_member = False # Try this
base_price = 100

discount_rate = 0.10 if is_member else 0.0
final_price = base_price * (1 - discount_rate)

print(f"Is member: {is_member}")
print(f"Discount rate: {discount_rate*100}%")
print(f"Final price: ${final_price}")
```

## Pitfall: order of `elif` statements

### Specificity matters

- Conditions in an `if...elif...else` chain are evaluated top to bottom, and the first `True` branch wins (the rest is skipped)
  - Therefore, you will often need to place **more specific conditions before more general ones**

Problematic order (grading):

```python
score = 95
grade = ""
# This is problematic because score >= 60 is true for 95
if score >= 60: 
    grade = "D" 
elif score >= 70:
    grade = "C"
elif score >= 80:
    grade = "B"
elif score >= 90:
    grade = "A" # This will never be reached if score is >= 60
else:
    grade = "F"
print(f"Score: {score}, Problematic Grade: {grade}")
```

Correct order (grading):

```python
score = 95 # Same score
grade = ""
# More specific conditions first
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(f"Score: {score}, Correct Grade: {grade}")
```

# A bit more about lists in Python 📝

## Lists with placeholder values

### Pre-allocating a list with `None`

- You can create a list of **placeholder values** using **`None`** (a **truly empty list** is just `[]`)
- `None` represents **the absence of a value** and is often used as a placeholder to be filled in later
- `None` is **not** the same as `0`, `False`, or an empty string `''`, and **you should never write it in quotes**
- Let's pre-fill a list with `None`:

```python
# Simply type "None"
list_answers = [None,None,None]
print(list_answers)
```

You can read more about `None` at <https://realpython.com/null-in-python/>.

## `is None` vs `== None`
### How to test for `None`

- `None` is a singleton; there is only ever one `None` object in memory
- Use **`is`** (and **`is not`**) to compare against it
- `==` works but is slower and is flagged by linters like `flake8` and `ruff`

```python
x = None

# Preferred:
if x is None:
    print("No value yet")
```

```python
# Works, but discouraged
if x == None:
    print("No value yet")
```

## Assigning or replacing values to lists

- You can assign or replace values in a list using the index of the element
- **The index of a list starts at 0**
- We use the following syntax:

```python
# What's the name of your hometown?
list_answers[0] = "Nashville"

print(list_answers)
```

## Finding values in a list

- You can find the index of a value in a list using the **`list.index()`** method
- This method returns the index of **the first occurrence** of the value in the list
- If the value is not in the list, it will raise a `ValueError`
- The syntax is as follows:

```python
cities = ["Nashville", "Bogota", "Atlanta"]
print(cities.index("Bogota"))
```

- You can also use the **`in`** operator to check if a value is in a list, as we saw before

```python
cities = ["Nashville", "Bogota", "Atlanta"]
"Bogota" in cities
```

- How to list _all_ values that match a search term?
- **Yes, there is a way**, but it's a bit more complicated
- We will learn about `for` loops soon, but for those who are curious... 🤓

```python
cities = ["Nashville", "Bogota", "Atlanta", "Nashville"]
matching_cities_loop = []

for index, city in enumerate(cities):
    if city.startswith("Nash"):
        matching_cities_loop.append((index, city))
print(matching_cities_loop)
```

## Common pitfalls with lists

- Lists and `np.array` use the same operators (`+`, `*`) but with **different meanings**
- On lists: `+` **concatenates**, `*` **repeats**; no element-wise maths
- On NumPy arrays: `+` and `*` work **element-by-element** (see lecture 03)

```python
# When you multiply a list times a number you repeat the list
list_a = [1,2,3]
print(list_a * 4)
```

```python
# When you add two lists, you concatenate them
list_b = [4,5,6]
print(list_a + list_b)
```

```python
# When you multiply an array times a number, you multiply each element
import numpy as np

vec_a = np.array(list_a)
print(vec_a * 4)
```
