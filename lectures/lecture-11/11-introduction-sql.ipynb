# Structured Query Language (SQL)

## SQL: A relational database management system (RDBMS)

- [SQL](https://en.wikipedia.org/wiki/SQL){data-modal-type="iframe"} stands for **Structured Query Language** and is used to interact with [relational databases](https://en.wikipedia.org/wiki/Relational_database){data-modal-type="iframe"}
- What is a **relational database**?
  - A database that stores data in tables and lets you define relationships between them
  - The most common type of database
  - Examples: PostgreSQL, MySQL, SQLite, Oracle, SQL Server
- Here we will focus on [SQLite](https://www.sqlite.org/index.html){data-modal-type="iframe"} 
- Why? It is open-source, powerful, and widely used in industry: phones, browsers, TVs, operating systems, and even cars use it!
- It has several advanced features too, such as [full-text search](https://www.sqlite.org/fts5.html){data-modal-type="iframe"} and [window functions](https://www.sqlite.org/windowfunctions.html){data-modal-type="iframe"}

## What is SQLite?

- SQLite is a **serverless database**: no server needed to run it
- It is **a single file** on your machine
- Pros: easy to install, lightweight, fast, great for small projects
- Cons: harder to scale, multiple writers don't play well together (though [LiteFS](https://fly.io/docs/litefs){data-modal-type="iframe"} helps), and time-series data is better served by [TimescaleDB for Postgres](https://www.timescale.com/)
- The [`sqlean` extensions](https://github.com/nalgeon/sqlean) cover many of the gaps

- A common myth: SQLite is "not a real database" because it lives in one file
- **This is false**: SQLite can handle [Exabyte-scale data](https://sqlite.org/hctree/doc/hctree/doc/hctree/index.html) (1M TB), far more than any classroom project will ever need
- The differences between SQL flavours are mostly small
- [Learn one and you can pick up the others quickly](https://www.postgresql.org/docs/current/sql-commands.html)

## Database structure and data types in SQLite

### Database structure

- The main object in SQL is the **table** (rows and columns)
- SQLite is **strongly typed**: each column has a specific data type
- Five core data types:
  - **`TEXT`**: strings (UTF-8)
  - **`INTEGER`**: whole numbers
  - **`NUMERIC`**: decimals (floating-point)
  - **`BLOB`**: binary data, stored as-is
  - **`NULL`**: missing or undefined values
- Plus extras like `REAL`, `BOOLEAN`, `DATE`, `TIME`, `DATETIME`

### Primary and foreign keys

- The **`PRIMARY KEY`** uniquely identifies each row, cannot be `NULL`, and is usually an integer
- Use **`AUTOINCREMENT`** to generate the values for you (`driver_id INTEGER PRIMARY KEY AUTOINCREMENT`)
- The **`FOREIGN KEY`** links two tables: a field in one table that refers to the `PRIMARY KEY` in another
- `ROWID` is a hidden auto-generated column, **but avoid it**: it can change after deletions and is not portable across SQL databases

# Table Schemas 📄

## What is a table schema?

- The [table schema](https://en.wikipedia.org/wiki/Database_schema){data-modal-type="iframe"} is the **structure of a table**: its columns, data types, constraints, and relationships
- Defined when the table is created, but can be modified later (though you usually won't need to)
- View it with the **`PRAGMA table_info()`** command

`PRAGMA table_info()` returns one row per column with:

- `cid`: column ID
- `name`: column name
- `type`: data type
- `notnull`: 1 if `NULL` is forbidden, 0 otherwise
- `dflt_value`: default value
- `pk`: 1 if part of the primary key, 0 otherwise

[![](figures/table-info.png){width="70%"}](#){data-modal-type="image" data-modal-url="figures/table-info.png"}

## Creating and modifying table schemas

- To create a table, use the **`CREATE TABLE`** command, as we saw before
- Use the `IF NOT EXISTS` clause to avoid errors when the table already exists
  - Example: `CREATE TABLE IF NOT EXISTS table_name (column1_name data_type, ...);`
- Use **`ALTER TABLE`** to modify an existing table:
  - Add a column: `ALTER TABLE table_name ADD column_name data_type;`
  - Drop a column: `ALTER TABLE table_name DROP COLUMN column_name;`
  - Rename a column: `ALTER TABLE table_name RENAME COLUMN old_name TO new_name;`

- To delete a table, use the **`DROP TABLE`** command
  - Example: `DROP TABLE table_name;`
  - Be careful: this deletes all the data in the table!
- To delete rows, use the **`DELETE FROM`** command
  - Example: `DELETE FROM table_name WHERE condition;`
  - Deletes all rows that satisfy the condition
  - Without a condition, it deletes every row in the table

## Groupby + Aggregate

### Groupby

- Use **`GROUP BY`** to group data by a column

```
SELECT nationality,
       SUM(victories) AS sum_victories,
       AVG(victories) AS mean_victories,
       MIN(victories) AS min_victories,
       MAX(victories) AS max_victories
FROM drivers
GROUP BY nationality;
```

![](figures/groupby.png)

### Round

- Use **`ROUND()`** to round the values
- The first argument is the value to round, the second is the number of decimal places
- The results match the previous slide, except that `mean_victories` is now rounded to one decimal place

```
SELECT nationality,
       SUM(victories) AS sum_victories,
       ROUND(AVG(victories), 1) AS mean_victories,
       MIN(victories) AS min_victories,
       MAX(victories) AS max_victories
FROM drivers
GROUP BY nationality;
```

![](figures/round.png)

## Filtering after aggregating with HAVING

- **`HAVING`** filters rows **after aggregating**
- **`WHERE`** filters rows **before aggregating**
- Position in the query: `HAVING` goes after `GROUP BY` and before `ORDER BY`

```
SELECT nationality,
       SUM(victories) AS sum_victories,
       ROUND(AVG(victories), 1) AS mean_victories,
       MIN(victories) AS min_victories,
       MAX(victories) AS max_victories
FROM drivers
GROUP BY nationality
HAVING SUM(victories) > 50;
```

## Python vs. SQL

- Common SQL commands and their pandas equivalents:

| SQL command   | Pandas equivalent                                                              |
|---------------|--------------------------------------------------------------------------------|
| `SELECT`      | `df[['column1', 'column2']]`                                                   |
| `INSERT INTO` | `pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)`                  |
| `WHERE`       | `df[df['column'] == value]`                                                    |
| `GROUP BY`    | `df.groupby('column')`                                                         |
| `HAVING`      | `df.groupby('column').filter(lambda x: condition)`                             |
| Aggregations  | `df.groupby('column').agg({'col1': 'sum', 'col2': 'mean'})`                    |

<br>

```
-- SQL
SELECT nationality,
       SUM(victories) AS sum_victories
FROM drivers
GROUP BY nationality
HAVING SUM(victories) > 10;
```

````python
# Python
import pandas as pd

# Group by 'nationality' and calculate 'sum_victories' using agg()
result = (
    drivers.groupby('nationality')
    .agg(sum_victories=('victories', 'sum'))
    .reset_index()
)

# Apply 'HAVING' condition
result = result[result['sum_victories'] > 10]
````
